```python
import pandas as pd

# 1. Path to your file
file_path = r'C:\Users\amoahj\Documents\KMR Submission 26\DiscoveryKMR.csv'
output_path = r'C:\Users\amoahj\Documents\KMR Submission 26\DiscoveryKMR_Updated.csv'

# 2. Load the file
df = pd.read_csv(file_path)

# 3. Process columns AD to AK (Index 29 to 36)
# We use iloc to select columns by position regardless of their names
cols_indices = list(range(29, 37)) 

for col_idx in cols_indices:
    col_name = df.columns[col_idx]
    
    # Convert to datetime - 'errors=coerce' handles empty cells
    # We use dayfirst=False because your example '10/30/2026' is Month/Day
    df[col_name] = pd.to_datetime(df[col_name], errors='coerce')
    
    # Convert to the specific string format "30-Oct-26"
    # Note: This turns the column into a String/Object so Excel is less likely to change it
    df[col_name] = df[col_name].dt.strftime('%d-%b-%y')

# 4. Save to Excel

# Save as Excel instead of CSV to preserve formatting
output_path_excel = r'C:\Users\amoahj\Documents\KMR Submission 26\DiscoveryKMR_Updated.xlsx'
df.to_excel(output_path_excel, index=False)
```
